```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tabla = pd.read_csv('MXvideos.csv', delimiter=',', encoding='iso-8859-1')
tabla.head(6)
```

```python
masv = tabla.sort_values(['views'], ascending=False).head(11)
masv
```

```python
tabla.shape
```

```python
canales = tabla[['channel_title', 'views']]
canales
```

```python
numeroview = canales.sort_values(['views'], ascending=False)
numeroview
```

```python
print(numeroview.groupby(['channel_title']).sum().sort_values(['views'], ascending=False))
```

```python
cmasvideos = canales['channel_title'].value_counts().sort_values(ascending=False).head(11)
cmasvideos
```

```python
cmasvideos.plot(kind='bar', color='blue')  # Puedes ajustar el número de bines
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Canales con mas VIDEOS')
plt.grid()
plt.show()
```

```python
tablik = tabla[['channel_title', 'likes']].sort_values(['likes'], ascending=False)
tablik
```

```python
tablik.head(11)
```

```python
tablik2 = tablik.groupby('channel_title').sum().sort_values(['likes'], ascending=False).head(11)
tablik2
```

```python
tablik2.plot(kind='bar', color='green')  # Puedes ajustar el número de bines
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Canales con mas LIKES')
plt.grid()
plt.show()
```

```python
tabdeslik = tabla[['channel_title', 'dislikes']].sort_values(['dislikes'], ascending=False)
tabdeslik
```

```python
tabla3 = tabdeslik.groupby(['channel_title']).sum().sort_values(['dislikes'], ascending=False).head(11)
tabla3
```

```python
tabla3.plot(kind='bar', color='red')  # Puedes ajustar el número de bines
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Canales con mas DESLIKES')
plt.grid()
plt.show()
```

```python
masvistos = tabla[['title', 'channel_title', 'views']].groupby(['title', 'channel_title']).sum().sort_values(['views'], ascending=False).head(11)
masvistos
```

```python
masvistos.plot(kind="bar", color="orange")
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('VIDEOS MAS VISTOS')
plt.grid()
plt.show()
```

```python
coment = tabla[['title', 'channel_title', 'comments_disabled']].groupby(['title', 'channel_title']).sum().sort_values(['comments_disabled'], ascending=False).head(11)
coment
```

```python
coment.plot(kind="bar", color="black")
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Videos con comentarios DESABILITADOS')
plt.grid()
plt.show()
```
